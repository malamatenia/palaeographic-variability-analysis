```python
#base imports
import os
from pathlib import Path
import rootutils
import sys
```

```python
root_path = Path(str(rootutils.find_root(search_from=os.getcwd(), indicator=".init_fr2813")))
sys.path.append(str(root_path / 'scripts' / 'utils'))
print(f"Root path: {root_path}")
from utils_prototypes import (
select_it, 
crop_it,
filter_it,
flag_it
)
```

```python
protos_dir = root_path / "results" / "prototypes" / "finetuned"

selected_protos_path = root_path / "results" / "prototypes" / "selected"

cropped_protos_path = root_path / "results" / "prototypes" / "cropped"

filtered_protos_path = root_path / "results" / "prototypes" / "filtered"

flagged_protos_path = root_path / "results" / "prototypes" / "flagged"
```

First, from the raw output of the finetuning, we select a specific range of characters that we'd like to process. Here we select characters from a-z, excluding, j,k,w.

```python
select_it(
    input_dir=str(protos_dir),
    output_dir=str(selected_protos_path),
    index_range_str="26-35,37-45, 47-49",
    ignore_files={'grid.png', 'grid-1l.png'}
)
```

```python
crop_it(
    input_dir=str(selected_protos_path),
    output_dir=str(cropped_protos_path),
    crop_top=7,
    crop_bottom=7,
    crop_left=7,
    crop_right=7
)
```

## Filter the prototypes

```python
filter_it(
    input_dir=cropped_protos_path,
    output_dir=filtered_protos_path,
    reference_dir=cropped_protos_path / "baseline",
    index_range=None,
    threshold_ratio=0.71
)
```

```python
flag_it(
    input_dir=filtered_protos_path,
    output_dir=flagged_protos_path,
    reference_dir=cropped_protos_path/ "baseline",
    index_range=None,
    default_threshold_ratio=0.71,
    sup_threshold_ratio=0.71,
    orange_threshold=(0.05, 0.2),
    red_threshold=0.2
)
```
